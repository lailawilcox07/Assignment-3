From these solutions, I learned that this problem follows a Fibonacci like sequence pattern. Following this to reach step n, you can come from either step n-1 (taking 1 step) or step n-2 (taking 2 steps). The total ways equals the sum of climbStairs(n-1) and climbStairs(n-2). I went with a recursive approach based on my flowchart.

```python
#Draft Code

def climbStairs(n):

    # Call ClimbStairs(n-1)
    result1 = climbStairs(n - 1)

    # Call ClimbStairs(n-2)
    result2 = climbStairs(n - 2)

    # Answer = result1 + result2
    answer = result1 + result2

    # Return answer
    return answer
```

```python
# Final working solution
# AI Disclaimer: Used ClaudeAI with prompt â€œCan you review my draft code and let me know what improvements can be made and why they would be improvements?â€ for understanding.
# AI Disclaimer: Used ClaudeAI with prompt "I am intrested in what memo is as I saw it in someone elses solution. Can you explain what it does and how it functions?"
# AI Disclaimer: Used ClaudeAI with prompt "I have implemented memo into my code can you check to see if I used it properly?"

def climbStairs(n):
    # Dictionary to store previously calculated results
    memo = {}

    def climb(n):
        # Base cases
        if n == 1:
            return 1

        if n == 2:
            return 2

        # Check if already calculated
        if n in memo:
            return memo[n]

        # Recursive calls
        memo[n] = climb(n - 1) + climb(n - 2)

        return memo[n]

    return climb(n)
```

From these solutions, I learned that there are multiple approaches to solve this problem. The first being a sorting approach where you sort both strings and compare them. The next being a hash table/dictionary approach where you count character frequencies using a dictionary. The last being an array counting approach where you use an array to count character frequencies for lowercase letters only.

I decided to implement the sorting approach because it's straightforward and directly relates to sorting concepts. The key insight is that if two strings are anagrams, they contain exactly the same characters with the same frequencies. When both strings are sorted, anagrams will produce identical sorted strings. I also learned that it's important to check if the lengths are equal first - if they're different, the strings can't be anagrams.

```python
#Draft Code

def isAnagram(s, t):
    # Check if lengths are different
    if len(s) != len(t):
        return False

    # Sort both strings
    sorted_s = sorted(s)
    sorted_t = sorted(t)

    # Compare sorted strings
    if sorted_s == sorted_t:
        return True
    else:
        return False
```

## 3. Final Correct Code



```python
# Final working solution

def isAnagram(s, t):
    # Check length and compare sorted strings
    return len(s) == len(t) and sorted(s) == sorted(t)
```
